```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Question 1
M = float(input("Enter a value for M: "))
x = np.linspace(-10,10,100)
y1 = M*x**2
y2 = M*np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'r--', label='y = M * x^2')
plt.plot(x, y2, 'b-', label='y = M * sin(x)')
plt.title("Mathematical Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
        'Score': [85, 90, 78, 88, 95]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.bar(df['Subject'], df['Score'], color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title("Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
for index, row in df.iterrows():
    plt.text(index, row.Score + 1, str(row.Score), ha='center', fontsize=12)
plt.grid(True,axis='y')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)
cumsum_data = np.cumsum(data)

fig, axes = plt.subplots(1,2,figsize=(15, 8))

axes[0].plot(cumsum_data, 'g-')
axes[0].set_title("Cumulative Sum Plot")
axes[0].grid()

axes[1].scatter(range(50), data, c='b', marker='o')
axes[1].set_title("Scatter Plot with Noise")
axes[1].grid()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url ="https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(url)
data.head()

sns.lineplot(x="month_number", y="total_profit", data=data)
plt.show()

for col in data.columns[1:6]:
  sns.lineplot(x="month_number", y=data[col],label = col, data=data)
plt.show()

for col in data.columns[1:6]:
  sns.barplot(x="month_number", y=data[col],label = col, data=data,alpha=0.4)
plt.show()
     
```
