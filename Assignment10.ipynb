```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
```

```python
texts = [
    "Love and hate are opposites that burn with the same fire. Love softens; hate hardens. Both can grow from deep emotion. It’s strange how love can turn into anger so quickly.",
    "Love and hate are powerful forces. Love connects; hate distorts. They both need passion. Maybe the true opposite of love isn’t hate—but indifference.",
    "The toaster launched toast. Kevin, sockless and holding peanut butter, stared. A squirrel watched from the window. Dinosaurs—or tacos—floated in the clouds. A ukulele played for no reason, but it worked."
]
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
bow = count_vectorizer.fit_transform(texts)
bow_df = pd.DataFrame(bow.toarray(), columns=count_vectorizer.get_feature_names_out())
print(bow_df)
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(texts)
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print(tfidf_df)
```

```python
for i, row in enumerate(tfidf_df.values):
    top_indices = row.argsort()[-3:][::-1]
    top_terms = [(tfidf_df.columns[idx], row[idx]) for idx in top_indices]
    print(f"\nText {i+1}:")
    for word, score in top_terms:
        print(f"  {word}: {score:.4f}")
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
text1 = """The MERN stack is widely used for building scalable, full-stack web applications with fast development cycles and a strong developer community."""

text2 = """AI has a vast scope across industries, enabling automation, smart decision-making, and personalized experiences in everything from healthcare to finance to entertainment."""
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text.split()
```

```python
tokens1 = preprocess(text1)
tokens2 = preprocess(text2)
```

```python
set1 = set(tokens1)
set2 = set(tokens2)
jaccard_sim = len(set1.intersection(set2)) / len(set1.union(set2))
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform([text1, text2])
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
```

```python
print(f"Jaccard Similarity: {jaccard_sim:.4f}")
print(f"Cosine Similarity: {cosine_sim:.4f}")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np
```

```python
paragraph = "The gentle rain fell softly on the windowpane, creating a soothing rhythm. Each drop seemed to whisper secrets of the quiet night. A feeling of calm settled over the room as the world outside faded into a peaceful hush. The scent of wet earth drifted in, carrying with it a sense of renewal. It was a moment of simple tranquility, a pause in the busyness of life."
tokenizer = Tokenizer()
tokenizer.fit_on_texts([paragraph])
total_words = len(tokenizer.word_index) + 1
```

```python
input_sequences = []
for line in paragraph.split('.'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

max_sequence_len = max([len(seq) for seq in input_sequences])
padded_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')
```

```python
x, y = padded_sequences[:,:-1], padded_sequences[:,-1]
y = np.eye(total_words)[y]
```

```python
model = Sequential([
    Embedding(total_words, 100, input_length=max_sequence_len-1),
    LSTM(150),
    Dense(total_words, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=100, verbose=0)
```

```python
def generate_text(seed_word, num_words):
    generated_text = seed_word
    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([seed_word])[0]
        padded_token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted_probabilities = model.predict(padded_token_list, verbose=0)[0]
        predicted_index = np.argmax(predicted_probabilities)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break
        generated_text += " " + output_word
        seed_word = output_word
    return generated_text
```

```python
seed_word = "rain"
num_lines_to_generate = 3
```

```python
for _ in range(num_lines_to_generate):
    generated_line = generate_text(seed_word, 8)
```
